```python
# Draw bounding boxes
def draw_boxes(image, bounds, color='green', width=4):
    im = PIL.Image.open(image)
    draw = ImageDraw.Draw(im)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return im

draw_boxes("/content/drive/MyDrive/OCR/Invoice/Handwritten.jpg", result)
```
